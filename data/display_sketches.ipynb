```python
import glob
import matplotlib.pyplot as plt
import os
import pandas as pd
from PIL import Image

#drawing_paths = glob.glob(os.getcwd()+'\\data\\drawings\\*\\*.png')
drawing_paths = [os.getcwd()+'\\drawings\\dragon\\0.png',
                 os.getcwd()+'\\drawings\\frown\\0.png',
                 os.getcwd()+'\\drawings\\hospital\\0.png',
                 os.getcwd()+'\\drawings\\island\\0.png',
                 os.getcwd()+'\\drawings\\king\\0.png',
                 os.getcwd()+'\\drawings\\apple\\1.png',
                 os.getcwd()+'\\drawings\\rabbit\\0.png',
                 os.getcwd()+'\\drawings\\thunderstorm\\0.png'
                 ]
drawing_classes = [path[path.index('drawings\\')+9:path.index('.png')-2] for path in drawing_paths]
drawing_dict = {"image_path":drawing_paths, "caption":drawing_classes}
data_frame = pd.DataFrame.from_dict(drawing_dict)
data_frame.head()
```

```python
fig, axs = plt.subplots(2, 4, dpi=300)
axs = axs.flatten()
for i, ax in enumerate(axs):
    path = data_frame["image_path"][i]
    prompt = data_frame["caption"][i]

    ax.imshow(Image.open(path))
    ax.set_title(prompt)
    ax.tick_params(labelleft=False, length=0)
```
